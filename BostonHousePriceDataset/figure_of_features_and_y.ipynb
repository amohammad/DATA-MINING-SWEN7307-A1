```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('BostonHousingDataset.csv', delimiter=',')
X_train = data[:,0:12]
X_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
y_train = data[:,13]

for i in range(13):
    plt.figure()
    plt.scatter(data[:,i],y_train)
    plt.xlabel('F#' + str(i+1) + ': ' + X_names[i])
    plt.ylabel("MEDV in $1000's")
    plt.show()
```
