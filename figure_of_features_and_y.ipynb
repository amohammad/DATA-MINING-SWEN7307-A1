```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('abalone.csv', delimiter=',')
X_train = data[:,0:6]
X_names = ['Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight' ]
y_train = data[:,7]

for i in range(7):
    plt.figure()
    plt.scatter(data[:,i],y_train)
    plt.xlabel('F#' + str(i+1) + ': ' + X_names[i])
    plt.ylabel('Rings, age in years')
    plt.show()
```
