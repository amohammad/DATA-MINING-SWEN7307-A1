```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('winequality-white.csv', delimiter=',')
X_train = data[:,0:10]
X_names = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']
y_train = data[:,11]

for i in range(11):
    plt.figure()
    plt.scatter(data[:,i],y_train)
    plt.xlabel('F#' + str(i+1) + ': ' + X_names[i])
    plt.ylabel("quality (score between 0 and 10)")
    plt.show()
```
